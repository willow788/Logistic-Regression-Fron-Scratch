```python
#writing the logistic regression
class logisticRegressionfromScratch:
    def  __init__(self, learning_rate = 0.01, iterations= 1000 l2=0.0):
        self.lr = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = 0
        self.cost_history = []
        self.l2 = l2

    def sigmoid(self, z):
        """using a sigmoid activation function"""
        z = np.clip(z, -500, 500)
        return 1 / (1 + np.exp(-z))


    def cost(self, h, y):
        """using cross entropy loss function"""
        m = len(y)
        eps= 1e -15
        h = np.clip(h, eps, 1-eps)
        base = - (1/m) * np.sum(y* np.log(h)) + (1 - y)* (np.log(1-h))
        if self.l2 and self.weights is not None:
            reg = (self.l2/ (2 * m)) * np.sum(self.weights ** 2)
            return base + reg
        return base

    

    def fit(self, X,y):
        """train model using  batch gradient descent"""
        X =  np.array(X)
        y = np.array(y).reshape(-1, 1)
        print("inside fit shapes:")
        print(X.shape)
        print(y.shape)
        
        m, n = X.shape
        self.weights = np.zeros((n,1))
        self.bias = 0
        self.cost_history = self.cost_history

        for _ in range(self.iterations):
            z = np.dot(X, self.weights) + self.bias
            h = self.sigmoid(z)

            dw = (1/m) * np.dot(X.T,(h-y))
            if self.l2:
                dw = dw+ (self.l2 / m)* self.weights
                
            db = 1/m * np.sum(h - y)
            self.bias -= self.lr* db
            self.weights -= self.lr * dw
            self.cost_history.append(self.cost(h, y))

    def predict(self, X):
        """make the predictions"""
        X = np.array(X)
        
        return (self.sigmoid(np.dot(X, self.weights) + self.bias)>= 0.5).astype(int)
    

        




    


```

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/breast-cancer-dataset/breast-cancer.csv")
df = df.copy()
print(df.head())

if 'id' in df.columns:
    df = df.drop(columns = ['id'])

df['diagnosis'] = df['diagnosis'].map({ 'M': 1, 'B': 0})

X = df.drop(columns= ['diagnosis'])
y = df['diagnosis']

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size= 0.25, shuffle= True
)
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)

y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Ensure labels are 1D numpy arrays
y_train_np = y_train.to_numpy()
y_test_np = y_test.to_numpy()
```

```python
print(X_train_scaled.shape, X_test_scaled.shape)

model = logisticRegressionfromScratch(learning_rate= 0.1, iterations= 1000)
model.fit(X_train_scaled, y_train_np)

predictions = model.predict(X_test_scaled)
accuracy = np.mean(predictions == y_test_np)
print(f"Model Accuracy:  {accuracy: .2f}")



```
